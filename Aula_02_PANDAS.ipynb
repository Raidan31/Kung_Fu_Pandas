```python
import pandas as pd
```

```python
#Leitura dos arquivos

df1 = pd.read_excel("/content/drive/MyDrive/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Salvador.xlsx")
```

```python
df5.head()
```

```python
#Juntando todas as planilhas

df = pd.concat([df1, df2, df3, df4 ,df5])
```

```python
#Exibindo as 5 primeiras linhas
df.head()
```

```python
#Pegar uma amostra aleatória do conjuto de dados
df.sample(5) #(5) numero de amostras 
```

```python
#Demonstra os tipos de dados da planilha
df.dtypes
```

```python
#Alterando o tipo de dado
df["LojaID"] = df["LojaID"].astype("object") 
#Transformou o LojaID de inteiro para objetc(String)
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Consultando linhas com valores faltantes

df.isnull().sum() #Se df for null traga a soma
```

```python
#Substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True) #mean == média e inplace = substituido na mémoria
```

```python
df["Vendas"].mean()
```

```python
#Substituindo os valores nulos por zero
df["Vendas"].fillna (0, inplace=True)
```

```python
#Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base em apenas 1 coluna (vendas)

df.dropna(subset = ["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how = "all", inplace=True)
```

```python
#Criando novas colunas

df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#Como encontrar a maior receita
df["Receita"].max()
```

```python
#Como encontrar a menor receita
df["Receita"].min()
```

```python
#Como retornar os maiores valores de uma determinada coluna  (no exemplo serão as 3 maiores receitas)
df.nlargest(3,"Receita")
```

```python
#Como retornar os menores valores de uma determinada coluna (no exemplo serão as 3 menores receitas)
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenar o conjunto de dados 
df.sort_values("Receita", ascending=False).head(10) 
#Vai ordenar com base na receita
#Ascending = False, para ser do maior para o menor
#head(10) - Para pegar as 10 maiores receitas
```
