```python
#Criando o dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas 
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Qual a receita total
df["Valor Venda"].sum()
```

```python
#Custo total
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #custo x quantidade
#Criando a coluna CUSTO
```

```python
df.head(1)
```

```python
#Qual o custo total?
round(df["Custo"].sum(),2)
#round serve para arredondar
#sum para somar todos os custos
```

```python
#Criar coluna de lucro (Valor da venda - custo)
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
round(df["Lucro"].sum(),2)
```

```python
#Criando uma coluna com o total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
#Agora informe a média de tempo de envio para cada MARCA, e para isso precisamos transormar a coluna Tempo_envio em numerica
#Extraindo apenas os dias da coluna tempo de envio (tá 20 days quero só os 20 sem o days)
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Agora está numerica 
df["Tempo_envio"].dtype
```

```python
#Médida de tempo de envio por marcas
round(df.groupby("Marca")["Tempo_envio"].mean(), 1)
```

```python
#Valores Ausentes
df.isnull().sum()
```

```python
#E se quisermos saber o lucro por ano e por marca?
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Melhorar a visualização tirando a notação cientifíca
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando Index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
#Serve pra sumarizar/agrupar a base e aplicar em outro excell
```

```python
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False ) 
```

```python
#Apresentando as informaçõess acima em grafico
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False ).plot.barh(title = "Total de produtos vendidos")
plt.xlabel("Total de Vendas")
plt.ylabel("Produtos");
```

```python
#Grafico lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title = "Lucro por Ano")
plt.xlabel("Ano")
plt.ylabel("Receitas");
```

```python
#Informações Acima
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Grafico com lucro por mês
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title = "Lucro por Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#Grafico lucro por Marca
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title = "Lucro por Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
#plt.xticks SERVE PARA MUDAR a observação(no caso das marcas em vez de estarem na vertical estão na horizontal) para horizontal
```

```python
#Grafico lucro por Classe
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title = "Lucro por Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```
